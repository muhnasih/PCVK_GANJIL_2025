```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Buat background hitam (500x500 pixel, 3 channel warna)
black_img2 = np.zeros((500, 500, 3), dtype=np.uint8)

# Definisi titik-titik polygon
vertices = np.array([[100,300],[200,200],[400,300],[200,400]], dtype=np.int32)
reshaped_vertices = vertices.reshape((-1,1,2))

# Gambar polygon
cv.polylines(black_img2, [reshaped_vertices], isClosed=True, color=(255, 0, 255), thickness=5)

# Konversi BGR -> RGB untuk matplotlib
img_rgb = cv.cvtColor(black_img2, cv.COLOR_BGR2RGB)

# Tampilkan
plt.imshow(img_rgb)
plt.axis("off")
plt.show()

```

2. Tampilkan image dalam channel Red-Blue dan Green-Blue saja!

```python
import cv2 as cv
import matplotlib.pyplot as plt

# Baca gambar
img = cv.imread('/content/drive/MyDrive/01.jpg')

if img is None:
    print("Gambar tidak ditemukan atau path salah")
else:
    # Konversi BGR -> RGB
    img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

    # Tampilkan gambar asli
    plt.imshow(img_rgb)
    plt.title("Gambar Asli")
    plt.axis('off')
    plt.show()

    # Red-Blue channel (hilangkan channel hijau)
    img_rb = img_rgb.copy()
    img_rb[:,:,1] = 0
    plt.imshow(img_rb)
    plt.title('Red-Blue Channel')
    plt.axis('off')
    plt.show()

    # Green-Blue channel (hilangkan channel merah)
    img_gb = img_rgb.copy()
    img_gb[:,:,0] = 0
    plt.imshow(img_gb)
    plt.title('Green-Blue Channel')
    plt.axis('off')
    plt.show()

```

3. Tampilkan image baris ke 20-115, kolom 25-120!


```python
import cv2 as cv
import matplotlib.pyplot as plt

# Baca gambar
img = cv.imread('/content/drive/MyDrive/01.jpg')

if img is None:
    print("Gambar tidak ditemukan atau path salah")
else:
    # Konversi BGR -> RGB
    img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

    # 1. Tampilkan gambar asli
    plt.imshow(img_rgb)
    plt.title("Gambar Asli")
    plt.axis('off')
    plt.show()

    # 2. Red-Blue channel (hilangkan channel hijau)
    img_rb = img_rgb.copy()
    img_rb[:,:,1] = 0
    plt.imshow(img_rb)
    plt.title('Red-Blue Channel')
    plt.axis('off')
    plt.show()

    # 3. Green-Blue channel (hilangkan channel merah)
    img_gb = img_rgb.copy()
    img_gb[:,:,0] = 0
    plt.imshow(img_gb)
    plt.title('Green-Blue Channel')
    plt.axis('off')
    plt.show()

    # 4. Image Slice (potong sebagian gambar)
    img_slice = img_rgb[20:116, 25:121]
    plt.imshow(img_slice)
    plt.title('Image Slice (Rows 20-115, Cols 25-120)')
    plt.axis('off')
    plt.show()

```
